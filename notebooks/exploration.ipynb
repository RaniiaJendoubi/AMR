```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("C:/Users/ranii/antimicrobial_resistance_project/Gambia data/gambia_antibiogram_sheet1.csv", sep=';')
```

```python
print("Aperçu des données :")
display(df.head())
```

```python
print("\nInfos générales :")
df.info()
```

```python
print("\nStatistiques descriptives :")
display(df.describe())
```

```python
print("\nValeurs manquantes par colonne :")
display(df.isnull().sum())
```

```python
if 'Organism' in df.columns:
    print("\nOrganismes uniques :")
    print(df['Organism'].unique())
if 'Antibiotic' in df.columns:
    print("\nAntibiotiques uniques :")
    print(df['Antibiotic'].unique())
if 'Resistance' in df.columns:
    print("\nValeurs de résistance :")
    print(df['Resistance'].unique())
```

```python
if 'Sample Type' in df.columns:
    plt.figure(figsize=(12,6))
    sns.countplot(y='Sample Type', data=df, order=df['Sample Type'].value_counts().index)
    plt.title("Distribution des échantillons par type")
    plt.xlabel("Nombre d'échantillons")
    plt.ylabel("Type d'échantillon")
    plt.show()
```

```python
le_organism = LabelEncoder()
le_antibiotic = LabelEncoder()
le_resistance = LabelEncoder()

if 'Organism' in df.columns:
    df['Organism_enc'] = le_organism.fit_transform(df['Organism'])
if 'Antibiotic' in df.columns:
    df['Antibiotic_enc'] = le_antibiotic.fit_transform(df['Antibiotic'])
if 'Resistance' in df.columns:
    df['Resistance_enc'] = le_resistance.fit_transform(df['Resistance'])

print("\nAperçu après encodage :")
display(df.head())
```

```python
df.to_csv("C:/Users/ranii/antimicrobial_resistance_project/Gambia data/gambia_antibiogram_sheet1.csv", index=False)
print("\nDataset prétraité sauvegardé sous : gambia_sheet1_preprocessed.csv")
```
